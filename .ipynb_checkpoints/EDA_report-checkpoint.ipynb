Client: Light rail network (Data and Client are confidential)

Problem Statement:

Business Problem: Track and switch maintenance is currently reactive which leads to service outages, increased community impact and significant expense due to replacement of parts that undergo accelerated degradation when operated in substandard condition. 

Proposed solution: Condition monitoring, if the areas of the network that are in need of maintenance can be detected before failure then preventative maintenance could be performed extending the lifetime of the network components and reducing the community impact and service outages.

Data Science Problem: Data of the network condition is collected weekly, however, there is currently no means of automated analysis, there is no temporal comparison. Consequently the X (usually 20) largest events are assessed by hand and the rest of the data is ignored. The client would like to assess track condition directly from the acceleration data without having to access the train control system to obtain train locations.

Data Science Solution: Build a model that integrates all data streams, identifies individual track locations and monitors the condition of those locations over time and flags any sections that are degraded past a given threshold.

This will take place in two stages, for the first stage we will build  a model that predicts the train locations, once that model is working a second model and further data will be used to build a model that assess track condition and identifies track sections in need of maintenance.


Data:

Data Description: 
The dataset is a single lap of the complete network. The primary data is obtained by measuring the vertical acceleration of each side of the front axle on a train. As the axel is one piece these measurements are only partially independent as vibration can be transmitted from one side through the axel. The data collection setup is completely separate from the train command and control system and is manually initiated before the automated test train is dispatched. The data recorder is retrieved after the test train completes the circuit of the track network. The secondary data is a GPS trace of the network during the test circuit. The labels are the event log files pertaining to the test train from the command and control network.

While the accelerometer data, GPS trace and event log have time stamps these are not synchronised and have substantially different temporal resolution. The event log data is reported to the nearest second while the accelerometer is reported to the nearest ten thousandth of a second. The acceleration data begins at 10:05 AM and goes to 13:19 PM for 194 minutes of data. The GPS trace has no universal time reference but goes for 11666079 milliseconds, or 194 minutes. The first logged event in the label stream occurs at 10:31:11AM and the last event occurs at 13:16:44 PM spanning 165 minutes.

No data cleaning was performed on the acceleration data. For the labels the data was filtered by the vehicle of interest (117) and on train integrity (Est) then all columns were discarded except time, actual velocity, track section, GW Area and the datetime. This network is speed limited to 90 km/hr with a normal high of ~82 k/hr so any speeds in excess of 90 km/hr were discarded. Finally, any duplicates or NaN were dropped.In this process ~16300 entries were discarded but the time span remains the same.

```python
print(df.shape)
dfi = df.loc[df["Veh ID"] == 117,:]
print(dfi.shape)
dfg = dfi.loc[dfi["Train Integrity"] == "Est",:]
print(dfg.shape)
df5 = dfg.loc[:,["Time", "Actual Velocity", "Track Section", "GW Area", "time"]]
df5 = df5.loc[df5["Actual Velocity"] <= 90, :]
df5 = df5.drop_duplicates()
df5 = df5.dropna()
print(df5.shape)
print(df5.head())
print(df5.tail())
print(df5.describe())
```

These plots demonstrated the need to condense the data to match the temporal resolution of the labels. The first approach I examined was short window fourier transforms (SWFT). The SWFT method is used to deconvolute the frequency components of each time period that make up the signal.

```python
df = np.log(pd.read_csv("ft_first_3000.csv", index_col=0))
print(df.info())
print(df.describe())
df.head()
```

```python
plt.figure(figsize=(8,8))
ax = sb.heatmap(df.T)
ax.invert_yaxis()
plt.show()
```

Frequency components over time of the first 3000 seconds of acceleration data. The stationary periods are clear (large dark bands). However it is also clear there there is some noise in the signal (fine horizontal bands that persist through the stationary phases) as some frequencies are elevated at all times, these are likely to be due to background mechanical systems on the train (pumps/fans)

The test circuit resembles a normal passage of a train throughout the circuit. In practice the test train is interleaved into normal service with the exception that the train doors do not open at the stations (no passengers on board the test train). Using the knowledge of the route it was possible to identify the start of the circuit at ~1330 seconds, the first normal station stop at ~1482 seconds and the first terminus at ~2489 seconds. Thus the first ~1300 seconds of acceleration data represents the train being moved throughout the staging yard in preparation of the test circuit. The terminus can be identified by the characteristic symmetry in the signals between 2460 and 2640 seconds as well as by counting stations. Plotting the initial section of the test circuit illustrates that there are speed specific features in the SWFT

```python
plt.figure(figsize=(10,8))
f = sb.heatmap(df.iloc[1300:2800,:].T, vmin=-10, vmax=2.2)
f.invert_yaxis()
f.set_title("Short window fourier transform of the first leg of test circuit")
f.set_xlabel("Time (s)")
f.set_ylabel("Frequency (Hz)")
plt.show()
f.figure.savefig("first_leg_a.png")
```

After examining the SWFT it is clear that the stationary periods are human detectable and should be amenable to classification. Furthermore there appears to be some speed dependent signal particularly in the 300-750 Hz band and to a lesser degree in the 750-1650 Hz band.

As a comparison a plot of the first 4000 velocity labels is provided. Due to the nature of the train control software the sampling rate is variable. It appears that the symmetrical pattern that occurs in the acceleration data centered at 2540 seconds occurs at ~10:47 in the labels.

```python
plt.figure(figsize=(10,8))
df5.iloc[:4000,:].plot(x="time", y= "Actual Velocity")
plt.show()
```

Focussing on the symmetrical period in both data and reducing the range of the colour axis to enhance contrast some correlation between frequency intensity and speed is apparent. As speed increases overall vibration increases but also the dominant frequency (the highest decibel frequency) increases. Also noticeable are some broadband spikes where all frequencies are suddenly elevated, it is thought that these may correspond impacts and are likely to be pronounced track defects.
